```python
import pandas as pd
```

```python
#insert name of OLDER csv file in the line below, within the quotations
file='june6at1005pmallcars.csv'
```

```python
#sets up and formats data
names =["web-scraper-order", "web-scraper-start-url", "listing links", "stocknumber","mileage","price", "Dealer", "Header", "Dealerindetails", "vin", "Dealership Full Name"]
data = pd.read_csv(file, names = names, delimiter=',')
raw_df = pd.DataFrame(data, columns=names)
old_car_df=raw_df.drop(columns=["Dealerindetails", "web-scraper-order", "listing links", "web-scraper-start-url", "web-scraper-order"])
old_car_df=old_car_df[["vin", "Header", "mileage", "price", "Dealer", "Dealership Full Name", "stocknumber"]]
```

```python
#displays older data
old_car_df
```

```python
#insert name of NEWER sheet here in the quotations
file='june13at1135pmallcars.csv'
```

```python
#sets up newer data
data = pd.read_csv(file, names = names, delimiter=',')
raw_df = pd.DataFrame(data, columns=names)
new_car_df=raw_df.drop(columns=["Dealerindetails", "web-scraper-order", "listing links", "web-scraper-start-url", "web-scraper-order"])
new_car_df=new_car_df[["vin", "Header", "mileage", "price", "Dealer", "Dealership Full Name", "stocknumber"]]
new_car_df
```

```python
#reads the vins from the dataframes into the old/new vins lists, and makes a copy of the newer vins list called new_cars 
old_vins=[]
new_vins=[]
for a in old_car_df['vin']:
    old_vins.append(a)
    
for a in new_car_df['vin']:
    new_vins.append(a)
    
new_cars=new_vins.copy()
```

```python
#checks each new vin to see if they were present in the older sheet - those that were are removed from the new_cars list
for n in new_vins:
    for o in old_vins:
        if n==o:
            new_cars.remove(o)
```

```python
#makes empty sold cars list, then checks each old vin to see if they're present in new vins list. 
#those that are not present are added to the sold_cars list
sold_cars=[]
for o in old_vins:
    if new_vins.count(o)==0:
        sold_cars.append(o)
```
